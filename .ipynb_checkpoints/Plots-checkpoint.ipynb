```python
#This function takes an array that has the positional (x,y,z) standard deviation of each cluster across the ligand series (10 ensembles)
#Output is a pml file that colors the most positionally variable clusters orange, and colors the most conserved clusters blue

active = [418,41,223,54,220,12,34,8,1,68,124,51,140,142,195,202,75,340] #cluster numbers < 5A to the active site

def stdDevArray(array):
    activeSite = []
    activeIndex = []
    sameRes = []
    sameResIndex = []
    diffRes =[]
    diffResIndex = []
 
    avg = np.average(array, axis=0)
    for num in range(array.shape[0]):  #0 to 420
        
        if num in active:
            activeSite.append(array[num])
            activeIndex.append(num)
            
        if array[num] > 1.32:       #25% most variable   
            diffRes.append(array[num])
            diffResIndex.append(num)
            
        elif array[num] < 0.62:     #25% least variable
       
            sameRes.append(array[num])
            sameResIndex.append(num)
            

    print(len(diffRes)) #check to see how many clusters are going to be colored orange 
    diffRes = np.column_stack(diffRes)
    print(len(sameRes)) #check to see how many clusters are going to be colored blue
    sameRes = np.column_stack(sameRes)
    activeSite = np.column_stack(activeSite)
    
    with open('spatialColoredWater.pml', 'w') as file:
        print('hide spheres', file=file)
        for element in sameResIndex:
            print('color blue, cluster_%s' %element, file=file)
            print('show spheres, cluster_%s' %element, file=file)
        for element in diffResIndex:
            print('color orange, cluster_%s' %element, file=file)
            print('show spheres, cluster_%s' %element, file=file)
        #for element in activeIndex:
           # print('color red, cluster_%s' %element, file=file)
            #print(element)
            #print('show spheres, cluster_%s' %element, file=file)

        #now size each sphere based on occupancy of that cluster 
        for element in range(420):
            for num in range(10):
                print('set sphere_scale, %s, cluster_%s and chain %s' %(waterArray[num][element], element, pdbs[num]), file=file)
            print('show spheres, cluster_%s' %element, file=file)
            
    file.close()
```

```python
#same idea as stdDevArray() but for rmsf. Input is rmsf array across the ligand series (10, totalNumberOfResidues)
#output is pml file coloring/displaying residues according to how much they vary by rmsf

def rmsfArray(array):
    sameRes = [] #consistently low rmsf
    no = [] #consistently high rmsf
    diffRes =[]  #changing/varying rmsf across the ligand series
    diffResIndex = []
    sameResIndex = []
    noIndex = []
    scoreStdDev = np.std(array, axis=0)
    avg = np.average(array, axis=0)
    for num in range(scoreStdDev.shape[0]):
 
        if scoreStdDev[num] > 0.04:      #25% most varying
            diffRes.append(array[:,num])
            diffResIndex.append(num)
        elif scoreStdDev[num] < 0.015: 
     
            if avg[num] < 3:             #if rmsf is less than 3A 
                sameRes.append(array[:,num])  #25% most consistently stable
                sameResIndex.append(num)
            else:
                no.append(array[:,num])
                noIndex.append(num)       #25% most consistently flexible
    
    print(len(diffRes))
    print(len(no))
    print(len(sameRes))
    #diffRes = np.column_stack(diffRes)
    #sameRes = np.column_stack(sameRes)
    
    with open('coloredProtein.pml', 'w') as file:
        print('hide sticks', file=file)
        for element in sameResIndex:
            if element < 168:
                print('color blue, resi %s and chain A' %element, file=file)
                print('show sticks, resi %s and chain A' %element, file=file)
            else:
                print('color blue, resi %s and chain B' %(element-167), file=file)
                print('show sticks, resi %s and chain B' %(element-167), file=file)
                
        for element in diffResIndex:
            if element < 168:
                print('color orange, resi %s and chain A' %element, file=file)
                print('show sticks, resi %s and chain A' %element, file=file)
            else:
                print('color orange, resi %s and chain B' %(element-167), file=file)
                print('show sticks, resi %s and chain B' %(element-167), file=file)
        for element in noIndex:
            if element < 168:
                print('color red, resi %s and chain A' %element, file=file)
                print('show sticks, resi %s and chain A' %element, file=file)
            else:
                print('color red, resi %s and chain B' %(element-167), file=file)
                print('show sticks, resi %s and chain B' %(element-167), file=file)
    file.close()
```

```python
#leiden cluster the water clusters!
#start with the (x,y,z) arrays of water cluster centers. 
#Goal is to cluster similar ensembles together based on their x,y,z water coordinates 

sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor='white')

#your array that you want to cluster is foobar
foobar = sc.AnnData(bigy) 
sc.pp.neighbors(foobar)
sc.tl.leiden(foobar, resolution=1)
sc.tl.umap(foobar)
sc.pl.umap(foobar, color='leiden')



```

```python
#takes the clustered array 
#outputs a dictionary: cluster index--> [PDBs in that cluster]
def clusterfcker(matrix, cluster_ref):
    maxval = 0
    for row in range(10):
        newval = int(cluster_ref.obs.leiden[row]) #has the cluster numbers 
        if newval > maxval:
            maxval = newval
    clusters = {}
    for y in range(maxval+1):
        clusters[y] = []

    for num in range(maxval+1):  #number of clusters 

        for x in range(10):

            if int(cluster_ref.obs.leiden[x]) == num:  #if this row is in the specific 'num' cluster, then
                    #clusters[num].append(matrix[x])
                    clusters[num].append(pdbs[x])

        clusters[num] = np.vstack(clusters[num])
        
    return clusters
```

```python
#outputs cool heatmaps with the most/least variable water clusters colored by cluster OCCUPANCY 
active = [418,41,223,54,220,12,34,8,1,68,124,51,140,142,195,202,75,340]
def stdDevArray(array):
    activeSite = []
    activeIndex = []
    sameRes = []
    diffRes =[]
    no = []
    noIndex = []
    diffResIndex = []
    sameResIndex = []
    scoreStdDev = np.std(array, axis=0)
    avg = np.average(array, axis=0)
    for num in range(scoreStdDev.shape[0]):
        
        if num in active:
            activeSite.append(array[:,num])
            activeIndex.append(num)
            
        if scoreStdDev[num] > 0.3:
            diffRes.append(array[:,num])
            diffResIndex.append(num)
        elif scoreStdDev[num] < 0.21:
        #if 10 > 9:
            if avg[num] > .72:  
                sameRes.append(array[:,num])
                sameResIndex.append(num)
            else:
                no.append(array[:,num])
                noIndex.append(num)
    
    print(len(diffRes))
    print(len(no))
    print(len(sameRes))
    diffRes = np.column_stack(diffRes)
    sameRes = np.column_stack(sameRes)
    activeSite = np.column_stack(activeSite)
    
    fig2, ax2 = plt.subplots(figsize=(15,3))

    im = ax2.imshow(diffRes, vmin=0, vmax=3,aspect="auto")
    ax2.set_title('Occupancy for Clusters with Highest Std Dev in Occupancy')
    ax2.set_xlabel("Cluster")
    ax2.set_yticks(np.arange(len(pdbs)))
    ax2.set_yticklabels(pdbs)
    ax2.set_xticks(np.arange(diffRes.shape[1]))
    ax2.set_xticklabels(diffResIndex)
    cbar = ax2.figure.colorbar(im, orientation='vertical', pad=0.05, fraction=0.05)
    cbar.ax.set_ylabel('Occupancy', rotation=-90, va="bottom")
    fig, ax = plt.subplots(figsize=(15,3))
    
    im = ax.imshow(sameRes, vmin=0, vmax=3, aspect="auto")
    ax.set_title('Occupancy for Clusters with Lowest Std Dev in Occupancy')
    ax.set_xlabel("Cluster")
    ax.set_yticks(np.arange(len(pdbs)))
    ax.set_yticklabels(pdbs)
    ax.set_xticks(np.arange(sameRes.shape[1]))
    ax.set_xticklabels(sameResIndex)
    cbar = ax.figure.colorbar(im, orientation='vertical', pad=0.05, fraction=0.05)
    cbar.ax.set_ylabel('Occupancy', rotation=-90, va="bottom")
    
    fig, ax = plt.subplots(figsize=(15,3))
    im = ax.imshow(activeSite, vmin=0, vmax=1, aspect="auto")
    ax.set_title('Occupancy for Clusters near Active Site')
    ax.set_xlabel("Cluster")
    ax.set_yticks(np.arange(len(pdbs)))
    ax.set_yticklabels(pdbs)
    ax.set_xticks(np.arange(activeSite.shape[1]))
    ax.set_xticklabels(activeIndex)
    cbar = ax.figure.colorbar(im, orientation='vertical', pad=0.05, fraction=0.05)
    cbar.ax.set_ylabel('Occupancy', rotation=-90, va="bottom")
```
